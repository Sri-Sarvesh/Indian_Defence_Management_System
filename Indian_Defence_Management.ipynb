```python
class DefenseSystem:
    def __init__(self, name, budget, personnel):
        self.name = name
        self.budget = budget
        self.personnel = personnel
        self.equipment = []

    def allocate_budget(self, amount):
        self.budget += amount
        print(f"{self.name} budget increased by {amount}.")

    def hire_personnel(self, personnel_count):
        self.personnel += personnel_count
        print(f"{self.name} hired {personnel_count} personnel.")

    def acquire_equipment(self, equipment):
        self.equipment.append(equipment)
        print(f"{self.name} acquired {equipment}.")


class Army(DefenseSystem):
    def __init__(self, name, budget, personnel, tanks, infantry):
        super().__init__(name, budget, personnel)
        self.tanks = tanks
        self.infantry = infantry

    def deploy_troops(self, location):
        print(f"Army troops deployed to {location}.")

    def conduct_military_exercise(self):
        print(f"{self.name} conducting a military exercise.")


class Navy(DefenseSystem):
    def __init__(self, name, budget, personnel, ships, submarines):
        super().__init__(name, budget, personnel)
        self.ships = ships
        self.submarines = submarines

    def patrol_seas(self, area):
        print(f"Navy patrolling the {area} seas.")

    def conduct_naval_drill(self):
        print(f"{self.name} conducting a naval drill.")


class AirForce(DefenseSystem):
    def __init__(self, name, budget, personnel, aircraft, squadrons):
        super().__init__(name, budget, personnel)
        self.aircraft = aircraft
        self.squadrons = squadrons

    def conduct_airstrike(self, target):
        print(f"Air Force conducted airstrike on {target}.")

    def organize_air_show(self):
        print(f"{self.name} organizing an air show.")


class IntelligenceAgency:
    def __init__(self, name, personnel, budget):
        self.name = name
        self.personnel = personnel
        self.budget = budget

    def gather_intelligence(self):
        print(f"{self.name} gathering intelligence.")

    def analyze_threats(self):
        print(f"{self.name} analyzing potential threats.")

    def conduct_covert_operations(self, target):
        print(f"{self.name} conducting covert operations on {target}.")


class CyberSecurityUnit:
    def __init__(self, name, personnel, budget):
        self.name = name
        self.personnel = personnel
        self.budget = budget

    def defend_against_cyber_attacks(self):
        print(f"{self.name} defending against cyber attacks.")

    def conduct_cyber_security_training(self):
        print(f"{self.name} conducting cyber security training programs.")
        
class MissileSystem:
    def __init__(self, name, range_km, quantity):
        self.name = name
        self.range_km = range_km
        self.quantity = quantity

    def launch_missile(self, target):
        print(f"{self.name} launched a missile at {target}.")

class RadarSystem:
    def __init__(self, name, coverage_range_km):
        self.name = name
        self.coverage_range_km = coverage_range_km

    def detect_threats(self):
        print(f"{self.name} detecting potential threats.")

class MedicalUnit:
    def __init__(self, name, personnel, budget, field_hospitals):
        self.name = name
        self.personnel = personnel
        self.budget = budget
        self.field_hospitals = field_hospitals

    def treat_casualties(self):
        print(f"{self.name} providing medical treatment to casualties.")

    def deploy_field_hospital(self, location):
        print(f"{self.name} deploying a field hospital to {location}.")

class LogisticsUnit:
    def __init__(self, name, personnel, budget, transport_vehicles):
        self.name = name
        self.personnel = personnel
        self.budget = budget
        self.transport_vehicles = transport_vehicles

    def supply_chain_management(self):
        print(f"{self.name} managing the supply chain.")

    def transport_materials(self, destination):
        print(f"{self.name} transporting materials to {destination}.")
        
class TrainingCenter:
    def __init__(self, name, personnel, training_courses):
        self.name = name
        self.personnel = personnel
        self.training_courses = training_courses

    def conduct_training(self, course, participants):
        print(f"{self.name} conducting {course} training for {participants} participants.")

class SpecialForcesUnit:
    def __init__(self, name, personnel, equipment, specializations):
        self.name = name
        self.personnel = personnel
        self.equipment = equipment
        self.specializations = specializations

    def execute_special_operations(self, target):
        print(f"{self.name} executing special operations on {target}.")

class CommunicationNetwork:
    def __init__(self, name, coverage_area, encryption_level):
        self.name = name
        self.coverage_area = coverage_area
        self.encryption_level = encryption_level

    def establish_secure_communication(self, participants):
        print(f"{self.name} establishing secure communication for {participants}.")

class AirDefenseSystem:
    def __init__(self, name, coverage_range_km, missile_system):
        self.name = name
        self.coverage_range_km = coverage_range_km
        self.missile_system = missile_system

    def intercept_aerial_threat(self, target):
        print(f"{self.name} intercepting aerial threat at {target}.")


# Example Usage
army = Army(name="Indian Army", budget=500000000, personnel=100000, tanks=2000, infantry=50000)
army.allocate_budget(100000000)
army.hire_personnel(5000)
army.acquire_equipment("Artillery")
army.deploy_troops("Kashmir")
army.conduct_military_exercise()

print(' ')

navy = Navy(name="Indian Navy", budget=300000000, personnel=50000, ships=50, submarines=10)
navy.allocate_budget(50000000)
navy.hire_personnel(2000)
navy.acquire_equipment("Submarine")
navy.patrol_seas("Arabian")
navy.conduct_naval_drill()

print(' ')

air_force = AirForce(name="Indian Air Force", budget=400000000, personnel=70000, aircraft=300, squadrons=20)
air_force.allocate_budget(80000000)
air_force.hire_personnel(3000)
air_force.acquire_equipment("Fighter Jet")
air_force.conduct_airstrike("Enemy Base")
air_force.organize_air_show()

print(' ')

intelligence_agency = IntelligenceAgency(name="RAW", personnel=5000, budget=200000000)
intelligence_agency.gather_intelligence()
intelligence_agency.analyze_threats()
intelligence_agency.conduct_covert_operations("Enemy Country")

print(' ')

cyber_security_unit = CyberSecurityUnit(name="Indian Cyber Command", personnel=1000, budget=100000000)
cyber_security_unit.defend_against_cyber_attacks()
cyber_security_unit.conduct_cyber_security_training()

print(' ')

missile_system = MissileSystem(name="Agni-V", range_km=5000, quantity=10)
missile_system.launch_missile("Enemy Territory")

print(' ')

radar_system = RadarSystem(name="Akash-1", coverage_range_km=300)
radar_system.detect_threats()

print(' ')

medical_unit = MedicalUnit(name="Armed Forces Medical Corps", personnel=2000, budget=100000000, field_hospitals=5)
medical_unit.treat_casualties()
medical_unit.deploy_field_hospital("Conflict Zone")

print(' ')

logistics_unit = LogisticsUnit(name="Army Logistics Command", personnel=3000, budget=150000000, transport_vehicles=1000)
logistics_unit.supply_chain_management()
logistics_unit.transport_materials("Forward Operating Base")

print(' ')

training_center = TrainingCenter(name="Defence Training Institute", personnel=100, training_courses=["Combat Training", "Intelligence Analysis"])
training_center.conduct_training("Combat Training", participants=50)

print(' ')

special_forces = SpecialForcesUnit(name="Para Commandos", personnel=200, equipment=["Night Vision Goggles", "Silenced Weapons"], specializations=["Counter-terrorism", "Reconnaissance"])
special_forces.execute_special_operations("Enemy Hideout")

print(' ')

communication_network = CommunicationNetwork(name="SecureCom", coverage_area="National", encryption_level="AES-256")
communication_network.establish_secure_communication(participants="Military Commanders")

print(' ')

air_defense_system = AirDefenseSystem(name="Akash-2", coverage_range_km=400, missile_system=MissileSystem(name="Barak", range_km=200, quantity=20))
air_defense_system.intercept_aerial_threat("Enemy Aircraft")
```
